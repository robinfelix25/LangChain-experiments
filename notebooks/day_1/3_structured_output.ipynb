```python
from langchain.chat_models import init_chat_model


llm_ollama = init_chat_model(model="ollama:gemma3:4b", temperature=0)
response = llm_ollama.invoke("Bro, tell me a fun fact")

response.content
```

```python
from langchain_core.prompts import ChatPromptTemplate

u_input = input("Enter topic : ")
d_prompt = ChatPromptTemplate.from_messages([
    ('system' , 'You are a helpful assistant. Return a structure output in key-value pair fun : fun , fact: fact'),
    ('human','tell me a fun fact {topic}. ')
])

r_prompt = d_prompt.invoke({'topic' : u_input})

llm_ollama.invoke(r_prompt.messages).content
```

## Using Pydantic method

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup : str = Field(description="The setup for the joke")
    punchline : str = Field(description="The Punchline for the joke")
```

```python
llm_Structured_output = llm_ollama.with_structured_output(llm_schema)
```

```python
result = llm_Structured_output.invoke("tell me a joke")
```

```python
result
```

```python
result.setup, result.punchline
```

## Using Typed Dict
- Similar to Pydantic model , But less strict means 'different key can be used'
for ex :  setup --> ketchup is also allowed 🤣

```python
from typing import TypedDict


class llm_schema_td(TypedDict):
    setup : str
    punchline : str
```

```python
llm_structured_out_td = llm_ollama.with_structured_output(llm_schema_td)
```

```python
result_td = llm_structured_out_td.invoke("tell me a joke")
```
