```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

## Chain with custom runnable

```python
#TASK -1 # create prompt


prompt_template = ChatPromptTemplate.from_messages([
    ('system', 'Yoy are an useful chat assistant'),
    ('human', '{user_input}')
])
```

```python
#TASK -2

llm = init_chat_model(model="ollama:gemma3:4b")
```

```python
#TASK - 3 :
str_parser = StrOutputParser()
```

```python
# Just to understand runnable
from langchain_core.runnables import RunnableLambda

def runnable_hello(text: str) -> str:
    return "hello" + text

run = RunnableLambda(runnable_hello)

run.invoke("Robin")
```

```python
# TASK  - 4 :
from langchain_core.runnables import RunnableLambda

def dictionary_maker(text : str) -> dict:
    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

```python
# TASK - 5 [Template for post]: 
from langchain_core.prompts import ChatPromptTemplate

prompt_post = ChatPromptTemplate.from_messages([
    ('system','You are a social media post generator'),
    ('human', 'Create a post for the following text for Linkedin: {text}')
])

```

```python
# TASK - 6 [Use the LLM]
```

```python
# TASK - 7 [StrOutParser]
```

```python
#TASK 8 [Remove \n\n]

def remove_newline(text: str) -> str:
    return text.replace("\n\n", "")

remove_newline_runnable = RunnableLambda(remove_newline)
```

```python
chain = prompt_template | llm | str_parser | dictionary_maker_runnable | prompt_post | llm | str_parser | remove_newline_runnable
```

```python
chain.invoke('what is the capital of france')
```
