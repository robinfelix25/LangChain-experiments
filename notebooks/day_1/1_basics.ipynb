```python
from langchain.chat_models import init_chat_model


llm_ollama = init_chat_model(model="ollama:gemma3:4b", temperature=0)
response = llm_ollama.invoke("Bro, tell me a fun fact")

response.content
```

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter your topic")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact {topic}")

ready_prompt = dynamic_prompt.invoke({'topic': user_input})

llm_ollama.invoke(ready_prompt).content
```
