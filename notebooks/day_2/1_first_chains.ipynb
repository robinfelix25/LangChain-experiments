```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
#TASK -1 # create prompt


prompt_template = ChatPromptTemplate.from_messages([
    ('system', 'Yoy are an useful chat assistant'),
    ('human', '{user_input}')
])
```

```python
#TASK -2

llm = init_chat_model(model="ollama:gemma3:4b")
```

```python
#TASK - 3 :
str_parser = StrOutputParser()
```

## method 1 : Manual Invocation (not recommended)

```python
template = prompt_template.invoke({'user_input' : 'What capital of USA'})
res = llm.invoke(template)
final_res = res.content
final_res
```

## method 2 : Chain Invocation (preferrable)
- Manual invocation is not the right way to do it
- Instead, use chain to connect all the blocks and run it

```python
chain = prompt_template | llm | str_parser

chain.invoke({'user_input' : 'Whats the capital of USA'})
```

## 3 : Runnable Sequence (optional)

```python
from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(prompt_template, llm, str_parser)
chain_1.invoke({'user_input' : 'Whats the capital of USA'})
```
